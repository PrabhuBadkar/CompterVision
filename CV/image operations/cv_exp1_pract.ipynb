##**Implement code to resize an image to half its original size.**

```python
image.shape
```

```python
height , width = image.shape[:2]
```

```python
height = int(height/2)
width = int(width/2)
```

```python
resized_img = cv2.resize(image,(height,width))
cv2_imshow(resized_img)
```

## **Write a program to rotate an image by 90 degrees.**

```python
rot_90_cw = cv2.rotate(resized_img,cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(rot_90_cw)
```

## **Modify the code to save the complemented image as a PNG file instead of JPG.**

```python
gray_img = cv2.cvtColor(resized_img,cv2.COLOR_BGR2GRAY)
comp_img = 255 - gray_img
cv2.imwrite('comp_png.png',comp_img)
cv2_imshow(comp_img)
```

## **Implement a function to overlay text on an image using OpenCV.**

```python
def text_over_img(image):
  img_cp = image.copy()
  font = cv2.FONT_HERSHEY_COMPLEX
  no_dup = cv2.LINE_AA
  text = input("Enter the text you want to overlay = ")
  font_size= int(input("Enter the size of text = "))
  thickness = int(input("Enter the thickness of text = "))
  x=int(input("Enter the x coordinate= "))
  y=int(input("Enter the y coordinate= "))
  position = (x,y)

  text_overlay = cv2.putText(img_cp,text,position,font,font_size,(255,0,0),thickness,no_dup)
  return text_overlay
```

```python
t = text_over_img(resized_img)
cv2_imshow(t)
```
